```python
import pandas as pd
import matplotlib.pyplot as plt

from xrayto3d_preprocess import scatterplot_1d, horizontal_line
```

```python
stats_path = 'rib_stats.csv'
df = pd.read_csv(stats_path)
```

```python
scatterplot_1d(df['total_voxels'].tolist(),label='rib')
plt.ylabel('#foreground-voxels')
plt.xlabel('Sample ID')
plt.legend(loc='upper left')
plt.savefig('totalsegmentator_rib_voxel_plot.pdf')
```

```python
df_non_empty = df[df['total_voxels'] != 0]
print('scans containing atleast some rib voxels',df_non_empty['total_voxels'].count())
print('scans containing atleast some voxels for all 12 ribs',df[df.transpose().all()]['total_voxels'].count())
```
